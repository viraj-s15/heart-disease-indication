```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
filename = "../../data/processed/heart_disease_processed.csv"
df = pd.read_csv(filename)
df.head()
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.head()
```

```python
correlation = df.corr()
correlation.style.background_gradient(cmap="coolwarm")
```

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.boxplot(data=df, width=0.6, whis=500, ax=ax, fliersize=5)
plt.xticks(rotation=90)
plt.title("Visualization of outliers")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(
    y=df["Age"],
    x=df["PhysicalActivity"],
    ax=ax[0],
    width=0.4,
)

sns.boxplot(
    y=df["BMI"],
    x=df["PhysicalActivity"],
    ax=ax[1],
    width=0.4,
)

plt.tight_layout()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(
    y=df["Diabetic"],
    x=df["PhysicalActivity"],
    ax=ax[0],
    whis=500,
    width=0.4,
)

sns.boxplot(
    y=df["GenHealth"],
    x=df["PhysicalActivity"],
    ax=ax[1],
    whis=500,
    width=0.4,
)

plt.tight_layout()
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(
    x="Age",
    hue="PhysicalActivity",
    data=df,
    palette="colorblind",
    edgecolor=sns.color_palette(as_cmap="RdGy_r", n_colors=5),
)

print("0 = Physical Activity", "1 = No Physical Activity")
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(
    x="SkinCancer",
    hue="PhysicalActivity",
    data=df,
    palette="colorblind",
    edgecolor=sns.color_palette(as_cmap="RdGy_r", n_colors=5),
)

print("0 = Physical Activity", "1 = No Physical Activity")
```
